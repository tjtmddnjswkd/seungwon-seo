```python
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn import datasets
X, y = datasets.make_classification(
n_samples=200, n_features=2, random_state=333,
n_informative=2, n_redundant=0, n_clusters_per_class=1)
print(X.shape)
print(y.shape)
```

```python
def sigmoid(s):
  return 1/(1+np.exp(-s))

def loss(y,h):
  return (-y*np.log(h) - (1-y)*np.log(1-h)).mean()

def gradient(X,y,w):
  return -(y*X)/(1+np.exp(-y*np.dot(w,X)))
```

```python
#GD
X_bias = np.append(np.ones((X.shape[0], 1)), X, axis=1)
y_g = np.array([[1] if label == 1 else [-1] for label in y])
y_l = np.array([[1] if label == 1 else [0] for label in y])
w = np.array([[np.random.uniform(-1, 1)] for _ in range(X.shape[1]+1)])

max_iter=100
learning_rate =0.03
threshold=0.5
for _ in range(max_iter):
  grad_E_in=0
  n=X_bias.shape[0]
  for i in range(X_bias.shape[0]):
    grad_E_in+=gradient(X_bias[i,:,np.newaxis],y_g[i,np.newaxis,:],w.T)
  w=w-learning_rate*(1/n)*grad_E_in

  probabilities=sigmoid(np.dot(X_bias,w))
  predictions=[[1] if p>threshold else [-1] for p in probabilities]

  print("loss : %.2f, accuracy : %.2f"%(loss(y_l,probabilities),accuracy_score(y_g,predictions)))
```

```python
#SGD
X_bias = np.append(np.ones((X.shape[0], 1)), X, axis=1)
y_g = np.array([[1] if label == 1 else [-1] for label in y])
y_l = np.array([[1] if label == 1 else [0] for label in y])
w = np.array([[np.random.uniform(-1, 1)] for _ in range(X.shape[1]+1)])

max_iter=100
learning_rate =0.03
threshold=0.5

for _ in range(max_iter):  
  
  r_num=np.random.randint(200)
  
  e_in=gradient(X_bias[r_num,:,np.newaxis],y_g[r_num,np.newaxis,:],w.T)
  w=w-learning_rate*e_in
  
  probabilities = sigmoid(np.dot(X_bias, w))
  predictions = [[1] if p > threshold else [-1] for p in probabilities]
    
  print("loss: %.2f, accuracy: %.2f"%(loss(y_l, probabilities), accuracy_score(y_g, predictions)))
```
